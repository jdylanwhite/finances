## Set Parameters

Set up the parameters we need to estimate retirement amounts. The parameters are:

* `startDate` : Date to start investment calculations
* `initialInvestment` : How much money we begin with prior to performing these calculations
* `retirementAmount` : The total investment goal that we'd like to reach before we decide to retire
* `monthlyInvestAmount` : The amount of money we'll invest each month
* `stockMarketAvg`: How well we expect the stock market to perform on average (without considering inflation)
* `stockMarketVar`: Variance for the stock market, to assess best/worst case scenarios
* `bondAvg`: As in `stockMarketAvg`, but for bond performance
* `bondVar`: As in `stockMarketVar`, but for bond performance
* `houseMarketAvg`: As in `stockMarketAvg`, but for housing performance
* `houseMarketVar`: As in `stockMarketVar`, but for housing performance

```r
# Set the start and end date in YYYY-MM-DD format
startDate = "2022-01-01"
endDate = "2042-01-01"

# Set the starting amount in dollars
initialStockInvestment = 0
initialBondInvestment = 0

# Set parameters for retirement in dollars
retirementAmount = 2000000
monthlyInvestAmount = 3500

# Set parameters for stock market performance
stockRateAvg = 0.1
stockRateVar = 0.02

# Set parameters for bond performance
bondRateAvg = 0.02
bondRateVar = 0.01

# Set parameters for house market peformance
houseRateAvg = 0.04
houseRateVar = 0.01

# Set stock/bond ratio
preRetirementStockPercentage = 0.9
postRetirementStockPercentage = 0.75

# Set parameters for inflation and adjust performances for inflation
inflation = 0.03
```

## Create Custom Functions

```r
elapsed_months <- function(startDate,endDate) {
    ed <- as.POSIXlt(endDate)
    sd <- as.POSIXlt(startDate)
    12 * (ed$year - sd$year) + (ed$mon - sd$mon)
}

four_percent_rule <- function(monthlyExpenses) {
    
    # Calculate the goal to meet the four percent rule
    goalAmount = (monthlyExpenses*12)/(0.04)
    
    return (goalAmount)
}

invest_for_n_months <- function(numMonths,monthlyInvestAmount,stockRateAvg,bondRateAvg,stockPercentage,stockRateVar=0.02,bondRateVar=0.01,initialStockInvestment=0,initialBondInvestment=0) {

    # Initialize total numbers
    stockTotal = initialStockInvestment
    stockTotalWorst = initialStockInvestment
    stockTotalBest = initialStockInvestment
    bondTotal = initialBondInvestment
    bondTotalWorst = initialBondInvestment
    bondTotalBest = initialBondInvestment

    # Initialize data frame
    df = data.frame(matrix(ncol=7,nrow=0))
    
    # Loop over months
    for (i in (1:numMonths)) {

        # Add compound interest for stock market investments
        stockTotal = stockTotal*(stockRateAvg/12.+1)
        stockTotalBest = stockTotalBest*((stockRateAvg+stockRateVar)/12.+1)
        stockTotalWorst = stockTotalWorst*((stockRateAvg-stockRateVar)/12.+1)
        
        # Add monthly contributions for stock market investments
        stockTotal = stockTotal + monthlyInvestAmount*stockPercentage
        stockTotalBest = stockTotalBest + monthlyInvestAmount*stockPercentage
        stockTotalWorst = stockTotalWorst + monthlyInvestAmount*stockPercentage

        # Add compound interest for bond investments
        bondTotal = bondTotal*(bondRateAvg/12.+1)
        bondTotalBest = bondTotalBest*((bondRateAvg+bondRateVar)/12.+1)
        bondTotalWorst = bondTotalWorst*((bondRateAvg-bondRateVar)/12.+1)

        # Add monthly contributions for bond investments
        bondTotal = bondTotal + monthlyInvestAmount*(1-stockPercentage)
        bondTotalBest = bondTotalBest + monthlyInvestAmount*(1-stockPercentage)
        bondTotalWorst = bondTotalWorst + monthlyInvestAmount*(1-stockPercentage)
        
        # Add row to the datafra,e
        row = (c(i,stockTotal,stockTotalWorst,stockTotalBest,bondTotal,bondTotalWorst,bondTotalBest))
        df = rbind(df,row);
        
    }
    
    # Give column names to the data frame
    colnames(df) <- c("month","stocks", "stocks, worst", "stocks, best", "bonds", "bonds, worst", "bonds, best")

    return (df)
    
}

invest_for_n_more_months <- function(numMonths,df,monthlyInvestAmount,stockRateAvg,bondRateAvg,stockPercentage,stockRateVar=0.02,bondRateVar=0.01) {
    
    # Initialize total numbers
    stockTotal = as.numeric(df[nrow(df),]['stocks'])
    stockTotalWorst = as.numeric(df[nrow(df),]['stocks, worst'])
    stockTotalBest = as.numeric(df[nrow(df),]['stocks, best'])
    bondTotal = as.numeric(df[nrow(df),]['bonds'])
    bondTotalWorst = as.numeric(df[nrow(df),]['bonds, worst'])
    bondTotalBest = as.numeric(df[nrow(df),]['bonds, best'])
    currentMonth = as.numeric(df[nrow(df),]['month'])
    
    # Loop over months
    for (i in (1:numMonths)) {

        # Add compound interest for stock market investments
        stockTotal = stockTotal*(stockRateAvg/12.+1)
        stockTotalBest = stockTotalBest*((stockRateAvg+stockRateVar)/12.+1)
        stockTotalWorst = stockTotalWorst*((stockRateAvg-stockRateVar)/12.+1)
        
        # Add monthly contributions for stock market investments
        stockTotal = stockTotal + monthlyInvestAmount*stockPercentage
        stockTotalBest = stockTotalBest + monthlyInvestAmount*stockPercentage
        stockTotalWorst = stockTotalWorst + monthlyInvestAmount*stockPercentage

        # Add compound interest for bond investments
        bondTotal = bondTotal*(bondRateAvg/12.+1)
        bondTotalBest = bondTotalBest*((bondRateAvg+bondRateVar)/12.+1)
        bondTotalWorst = bondTotalWorst*((bondRateAvg-bondRateVar)/12.+1)

        # Add monthly contributions for bond investments
        bondTotal = bondTotal + monthlyInvestAmount*(1-stockPercentage)
        bondTotalBest = bondTotalBest + monthlyInvestAmount*(1-stockPercentage)
        bondTotalWorst = bondTotalWorst + monthlyInvestAmount*(1-stockPercentage)
        
        # Increment month count
        currentMonth = currentMonth+1
        
        # Add row to the datafra,e
        row = (c(currentMonth,stockTotal,stockTotalWorst,stockTotalBest,bondTotal,bondTotalWorst,bondTotalBest))
        df = rbind(df,row);
        
    }
    
    # Give column names to the data frame
    colnames(df) <- c("month","stocks", "stocks, worst", "stocks, best", "bonds", "bonds, worst", "bonds, best")

    return (df)
    
}

invest_until_x_amount <- function(goalAmount,monthlyInvestAmount,stockRateAvg,bondRateAvg,stockPercentage,stockRateVar=0.02,bondRateVar=0.01,initialStockInvestment=0,initialBondInvestment=0) {

    # Initialize total numbers
    stockTotal = initialStockInvestment
    stockTotalWorst = initialStockInvestment
    stockTotalBest = initialStockInvestment
    bondTotal = initialBondInvestment
    bondTotalWorst = initialBondInvestment
    bondTotalBest = initialBondInvestment
    currentMonth = 0
    
    # Initialize data frame
    df = data.frame(matrix(ncol=7,nrow=0))
        
    # Loop over months
    while (stockTotal+bondTotal < goalAmount) {

        # Advance current month
        currentMonth = currentMonth+1
        
        # Add compound interest for stock market investments
        stockTotal = stockTotal*(stockRateAvg/12.+1)
        stockTotalBest = stockTotalBest*((stockRateAvg+stockRateVar)/12.+1)
        stockTotalWorst = stockTotalWorst*((stockRateAvg-stockRateVar)/12.+1)
        
        # Add monthly contributions for stock market investments
        stockTotal = stockTotal + monthlyInvestAmount*stockPercentage
        stockTotalBest = stockTotalBest + monthlyInvestAmount*stockPercentage
        stockTotalWorst = stockTotalWorst + monthlyInvestAmount*stockPercentage

        # Add compound interest for bond investments
        bondTotal = bondTotal*(bondRateAvg/12.+1)
        bondTotalBest = bondTotalBest*((bondRateAvg+bondRateVar)/12.+1)
        bondTotalWorst = bondTotalWorst*((bondRateAvg-bondRateVar)/12.+1)

        # Add monthly contributions for bond investments
        bondTotal = bondTotal + monthlyInvestAmount*(1-stockPercentage)
        bondTotalBest = bondTotalBest + monthlyInvestAmount*(1-stockPercentage)
        bondTotalWorst = bondTotalWorst + monthlyInvestAmount*(1-stockPercentage)
        
        # Add row to the datafra,e
        row = (c(currentMonth,stockTotal,stockTotalWorst,stockTotalBest,bondTotal,bondTotalWorst,bondTotalBest))
        df = rbind(df,row);
        
    }
    
    # Give column names to the data frame
    colnames(df) <- c("month","stocks", "stocks, worst", "stocks, best", "bonds", "bonds, worst", "bonds, best")

    return (df)
    
}

get_total_investments <- function(df) {
    
    # Get total (bonds+stocks) and add new columns
    df['total'] = df['stocks']+df['bonds']
    df['total, worst'] = df['stocks, worst']+df['bonds, worst']
    df['total, best'] = df['stocks, best']+df['bonds, best']
    
    return (df)
    
}
```

```r
numMonths = elapsed_months(startDate,endDate) 
df = invest_for_n_months(numMonths,monthlyInvestAmount,stockRateAvg,bondRateAvg,preRetirementStockPercentage)
df = invest_for_n_more_months(numMonths,df,0,stockRateAvg,bondRateAvg,postRetirementStockPercentage)
df = get_total_investments(df)
```
